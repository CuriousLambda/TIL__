```python
import tensorflow as tf
import numpy as np
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity="all"
```

```python
input_data = np.array([ [1, 1], [4, 3] ])
target_data = np.array([ 0, 1 ])
```

```python
input_nodes = 2    # 입력노드 
hidden_nodes = 2   # 은닉노드 
output_nodes = 1   # 출력노드
```

```python
W2 = np.random.rand(input_nodes, hidden_nodes) 
b2 = np.random.rand(hidden_nodes) 
W3 = np.random.rand(hidden_nodes, output_nodes) 
b3 = np.random.rand(output_nodes) 
print("W2 = ", W2, ", W2.shape = ", W2.shape) 
print("b2 = ", b2, ", b2.shape = ", b2.shape) 
print("W3 = ", W3, ", W3.shape = ", W3.shape) 
print("b3 = ", b3, ", b3.shape = ", b3.shape)
```

```python
from tensorflow.keras.models import Sequential # 순차모델 생성하겠다.
from tensorflow.keras.layers import Dense

data = input_data
labels = target_data

# 모델 구성하기
model = Sequential()

# 입력 레이어와 히든 레이어를 쌓는 단순한 모델
model.add(Dense(1, input_dim = 2, activation = 'relu'))

model.add(Dense(2, activation = 'softmax'))
```

```python
model.compile('rmsprop', 'mse', metrics = ['acc'])
```

```python
model.fit(data, labels, epochs = 200)
```

```python
#7. 데이터 예측

test = np.random.rand(200,2)
predict = np.argmax(model.predict(test), axis = 1)
real = (np.sum(test, axis = 1) > 2.5) *1
print(sum(predict ==real) / 200.0)
```

```python
from tensorflow.keras.models import Sequential # 순차모델 생성하겠다.
from tensorflow.keras.layers import Dense


input_data = np.array([[1,2]])
target_data = np.array([1])

data = input_data
labels = target_data

# 모델 구성하기
model = Sequential()

# 입력 레이어와 히든 레이어를 쌓는 단순한 모델
model.add(Dense(1, input_dim = 2, activation = 'relu'))

model.add(Dense(3, activation = 'sigmoid'))
```

```python
model.compile('rmsprop', 'mse', metrics = ['acc'])
```

```python
model.fit(data, labels, epochs = 200)
```

```python
#7. 데이터 예측

test = np.random.rand(200,2)
predict = np.argmax(model.predict(test), axis = 1)
real = (np.sum(test, axis = 1) > 2.5) *1
print(sum(predict ==real) / 200.0)
```
