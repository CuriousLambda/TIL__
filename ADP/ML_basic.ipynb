### 머신러닝  : 기계학습  
> "The  field of study that  gives computers the ability to learn without being explicitly programmed ". <br/>
>  사람이 명시적으로  행동을 지시하지 않고 컴퓨터에게 학습 능력을 주는 것      

- 빅데이터를 통해서  규칙을 학습하게 하는 방식 
- 기계 모형이 학습하는 방법(methods)을 따라 데이터로 부터 규칙(rule)을 학습 하는 것  
    - 학습한 규칙  = 표현  = 예측 방정식  = 예측 함수  = 근사 함수  

###  머신 러닝의 종류
    1. 지도학습
        - 정답(올바른 결과 )주고  데이터를 기계에 학습 시키는 방법  
        - 회귀 데이터를 입력하면 출력으로 숫자를 반환 하는 방법 :예측 ex) 주가 예측 
        - 분류 데이터를 입력하면 출력으로 데이터의 속성이나 종류를 반환 하는 방법 : 라벨링 ex) 메일스팸감지)

    2. 비지도학습
        - 정답을 줄수 없는  방대한 데이터로 부터 자동적으로 산출한 특징량으로 부터 구조나, 경향, 법칙 등을  이용한 기계에 학습 시키는 방법 
        - 클러스터링 (데이터를 입력할 때 해당 데이터의 그룹화 결과를 반환 하는 방법)
        
    3. 강화학습
        - 스스로 시행 착오을 일으켜서 최적의 행동을 찾는 학습으로 최근의 목표를 달성하고 보상을 주는 것으로 향상하는 방법.

### 회귀 
1. 선형 회귀
    - 선형 회귀란, 데이터의 분포가 있을 때, 그 데이터에 제일많이  적용되는 직선을 구하는 분석 방법의 일종. 
    - 회귀 문제를 해결하는 가장 간단한 모델로  매출 예측이나 보험회사의 리스크 평가 등의 
    - 비즈니스 이용을 비롯해 스포츠 분석이나 환경 과학 등 다분야에서 활용

2. 로지스틱 회귀 
    - 분류 문제를 해결하는 모델, 여러 변수에서 특정 이벤트가 발생할 확률을 예측

### Q5) 생각해보자
[1차 결론] : 정답과 예측치의 오차의 단순함을 모두 0으로 만들면 판단이 모호하다.

[2차 결론] : 음수의 값으로 변환되는 것이 있기 때문이다.

[해결책] : 오차의 결과를 양수로 만들어 주는 작업을 새롭게 하자. 오차들의 제곱들의 평균을 구한 후 이 값의 제곱근으로 오차를 정의하자.
     <br/>   --> 추정된 선으로부터 점들이 평균적으로 얼마나 떨어져있는지를 판단한다.
     <br/>   --> 오차 제곱합을 최소화 하는 회귀계수

RMSE = Root Mean Square Error  
        
$$
RMSE = \sqrt{\frac{1}{n}\sum_{i=1}^{n}{(Y_i - f(X_i))^2}}
$$  

## 비용함수 머신러닝에서 사용하는 용어 위에랑 똑같은거임

Cost Function
$$
f(X) = \beta_1X + \beta_0
$$
$$
Cost(\beta_0, \beta_1) = \sqrt{\frac{1}{n}\sum_{i=1}^{n}{(Y_i - f(X_i))^2}}
$$

```python
def RMSE(a, b, X, y):
    RMSE = np.sqrt(((y - (a*X + b)) **2).mean())
    return RMSE
```

```python
# 첫번째 직선 f(x) = 2x+5
print(RMSE(2, 5, X, Y))
# 두번째 직선 f(x) = 30
print(RMSE(0, 30, X, Y))
# 세번째 직선 f(x) = -2x +55
print(RMSE(-2, 55, X, Y))
```

```python
from sklearn.metrics import mean_squared_error, r2_score

def rmse(y_true, y_pred):
    rems = np.sqrt(mean_squared_error(y_true, y_pred))
    print('rmse', rmse)
    return rmse
```

```python
# Code source: Jaques Grobler
# License: BSD 3 clause

import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error, r2_score

# Load the diabetes dataset
diabetes_X, diabetes_y = datasets.load_diabetes(return_X_y=True)

# Use only one feature
diabetes_X = diabetes_X[:, np.newaxis, 2]

# Split the data into training/testing sets
diabetes_X_train = diabetes_X[:-20]
diabetes_X_test = diabetes_X[-20:]

# Split the targets into training/testing sets
diabetes_y_train = diabetes_y[:-20]
diabetes_y_test = diabetes_y[-20:]

# Create linear regression object
regr = linear_model.LinearRegression()

# Train the model using the training sets
regr.fit(diabetes_X_train, diabetes_y_train)

# Make predictions using the testing set
diabetes_y_pred = regr.predict(diabetes_X_test)

# The coefficients
print("Coefficients: \n", regr.coef_)
# The mean squared error
print("Mean squared error: %.2f" % mean_squared_error(diabetes_y_test, diabetes_y_pred))
# The coefficient of determination: 1 is perfect prediction
print("Coefficient of determination: %.2f" % r2_score(diabetes_y_test, diabetes_y_pred))

# Plot outputs
plt.scatter(diabetes_X_test, diabetes_y_test, color="black")
plt.plot(diabetes_X_test, diabetes_y_pred, color="blue", linewidth=3)

plt.xticks(())
plt.yticks(())

plt.show()
```

MSE ( Mean Squared Error ) = 오차 제곱 평균 = 연속형변수값을 예측할 때 사용 : 다중분류시 사용

MAE (Mean Absolute Error, 평균 절대 오차)

$$
MAE(\hat{y}) := \frac{1}{N}\sum_{i=1}^{N} \left| y_i - \hat{y}_{i} \right|
$$


\begin{align} 
MSE(\hat{y}_{\theta_1}) &> MSE(\hat{y}_{\theta_2}) \ 
\Leftrightarrow ~~ RMSE(\hat{y}_{\theta_1})   &>   RMSE(\hat{y}_{\theta_2}) \ 
\Leftrightarrow ~~ \hspace{1.5em} {R^{2}}(\hat{y}_{\theta_1})   &<   {R^{2}}(\hat{y}_{\theta_2}) 
\end{align}

 

 -결정계수(R2)  : Y의 총 변동량 중에서 X에 의해서 설명된 량으로 '표본에 대한 회귀모형의 설명'을 말한다.  

   - 값이 클수록(1에 가까울수록) 모형이 적합하다고 평가할수 있다. 

   - 즉  ,  단순히 Y의 평균값을 사용했을 때 대비 X 정보를 사용함으로써 얻는 성능 향상을 갖는다. 

   - X 변수가 Y 변수의 분산을 얼마나 줄였는가에 대한 정도를 확인 할 때 사용한다.
   
   - R2 = 1 -> 모든 측정값들이 회귀직선에 있을 경우
   - R2 = 0 -> X,Y의 관계를 알 수 없다.
