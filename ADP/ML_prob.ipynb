# 추측통계의 기본

- 추측 통계: 일부 데이터로부터 전체의 통계적 성질을 추측


## 모집단과 표본


- 모집단(population) 
    - 조사 대상이 되는 관측 가능한 개체로 된 집단 전체
    
    
- 표본(sample)
    - 모집단에서 선택된 모집단 구성단위의 일부


- 표본 조사의 단계
    - 목표 모집단 -> 표본 프레임 -> 표본 단위
    - (Tp               SF                )

**전수조사와 표본조사**
- 전수조사
    - 모집단 전체를 조사
- 표본조사
    - 모집단의 일부를 조사

- 모수와 통계량
    - 모수(parameter)
        - 모집단의 특성을 나타내는 수치
    - (표본)통계량(sample statistic)
        - 표본의 특성을 나타내는 수치


### 표본조사 방식

1. 모집단 전체에서 필요한 크기만 추출
2. 모집단을 층으로 나누어 각 계층에서 추출
3. 모집단을 세밀하게 나눈 후 몇개 부분만 추출

- 확률 추출 : 모집단을 구성하는 객체가 표본에 포함될 확률을 미리 정한다.
- 목적 : 확률 추출을 하면 확률 이론을 바탕으로 표본 오차를 미리 추정하는 것이 가능하다.

> 표본오차 = 모집단의 모수의 참값과 표본 통계량 값의 차이

### 표본의 추출방법

- 무작위 추출(임의 추출): 임의로 표본을 추출하는 방법
- 복원추출: 여러 차례 동일한 표본을 선택하는 방법
- 비복원추출: 동일한 표본은 한 번만 선택하는 방법

## 확률 모델

### 확률의 기본

- 확률(probability)
    - 어떤 사건(event)이 일어날 가능성
    - 예. 
        - 동전 던지기에서 앞면이 나올 가능성
        - 주사위 던지기에서 특정 눈금이 나올 확률
        - 주식투자에서 이득을 볼 가능성
        
        

- 의사결정
    - 확실성(certainty) 보다는 불확실한 상황에서 이루어지는 경우가 많음
    - 미래에 대한 불확실성의 정도가 더욱 심하다고 할 수 있음 
    - 불확실성 하에서 의사결정의 오류를 줄이기 위해 확률에 대한 연구 필요


### 확률변수 : 취하는 값과 그 값이 나올 확률이 결정되어 있는 것
    - 일정한 확률로 나타나는(발생하는) 사건에 대해 숫자를 부여한 변수
    - ex. 주사위를 굴리면 1-6사이의 숫자가 약 1/6 확률로 발생

- 시행
    - 확률 변수의 결과를 관측하는 것
    - 주사위를 굴리면 어떤 숫자가 나올지 모르지만 숫자는 하나씩 확정
- 실현값
    - 시행에 의해 관측되는 값
    - 주사위를 굴려서 나온 값
- 사건
    - 시행 결과로 나타날 수 있는 일
    - 눈은 1 또는 눈은 홀수 등
    - 주사위의 눈은 확률 변수 X
        - 눈이 1이 되는 사건의 확률 P(x=1)=1/6
        - 눈이 홀수인 사건의 확률 P((X=1)U(X=3)U(X=5) = 1/6+1/6+1/6 = 1/2

    - 근원사건
        - 세부적으로 더 분해할 수 없는 사건
        
        
- 상호배반 : 동시에 일어날 수 없는 사건
    - 동전을 던져 앞면이 나왔을때 동시에 뒷변이 나오는 사건

- np.random.choice(a, size=None, replace=True, p=None)

    - a : 배열이면 배열 자체, 정수이면 arange(a)명령으로 배열 생성
    - size : 정수, 샘플 숫자
    - replace : boolean. True이면 복원추출, False면 비복원추출
    - p : 배열, 각 데이터가 선택될 수 있는 확률

```python
#1. 주사위를 10번 던져서 눈을 출력해보자.
import random

dice = [1, 2, 3, 4, 5, 6]

for i in range(1, 11):
    print(i, '번째는 ', random.choice(dice))
```

```python
#2. 주사위를 10번 던져서 눈이 3이 나오는 횟수를 출력해보자.

dice = [1, 2, 3, 4, 5, 6]

cnt = 0

for i in range(1, 11):
    res = random.choice(dice)
    if res == 3:
        cnt += 1

print(cnt)
    
```

```python
#3. 주사위를 1000번 던져서 눈이 5가 나올 수 있는 확률을 구하세요.

dice = [1, 2, 3, 4, 5, 6]

cnt = 0

for i in range(1, 1001):
    res = random.choice(dice)
    if res == 5:
        cnt += 1

print(cnt/1000)
```

```python
#4. 주사위를 10번 던져서 짝수가 나오는 횟수를 구해보자.

dice = [1, 2, 3, 4, 5, 6]

cnt = 0

for i in range(1, 11):
    res = random.choice(dice)
    if res%2 == 0:
        cnt += 1

print(cnt)
```

```python
#5. 주사위를 10번 던졌을 때 짝수가 나오는 횟수를 출력하는 것을 5번 반복하자.

dice = [1, 2, 3, 4, 5, 6]

for j in range(1,6):
    cnt = 0
    for i in range(1, 11):
        res = random.choice(dice)
        if res%2 == 0:
            cnt += 1

    print(cnt)
```

```python
#6. 한개의 주사위를 360번 던져서 3의 배수의 눈이 나오는 횟수를 출력하는 것을 1000번 반복하고
#   그 1000개를 X변수에 대입해서 X의 평균, 분산, 표준편차를 구하자.

dice = [1, 2, 3, 4, 5, 6]

x = []

for j in range(1,1001):
    cnt = 0
    for i in range(1, 360):
        res = random.choice(dice)
        if res%3 == 0:
            cnt += 1
    x.append(cnt)
    
print(np.mean(x))
print(np.var(x))
print(np.std(x))
```

```python
#7. 주사위 2개를 동시에 던져서 두개의 주사위의 합을 구하는 실행문을 1000번 반복해서
#   1000개를 a변수 리스트 객체에 대입하자. (while)

dice = [1, 2, 3, 4, 5, 6]
dice2 = [1, 2, 3, 4, 5, 6]
a = []

cnt = 0

while cnt < 1000:
    res2 = random.choice(dice2)
    res = random.choice(dice)
    a.append(res + res2)
    cnt = cnt + 1
print(len(a))
```

```python
#8. 7번의 문제에서 두개의 주사위의 눈의 합이 10이 되는 확률을 구하시오.

dice = [1, 2, 3, 4, 5, 6]
dice2 = [1, 2, 3, 4, 5, 6]
a = []
cnt = 0

while cnt < 1000:
    res2 = random.choice(dice2)
    res = random.choice(dice)
    
    if res2 + res == 10:
        a.append(res+ res2)
    cnt = cnt+1

print(len(a)/1000)
```

```python
#9. 주사위 한개를 288회 던져서 주사위의 눈이 5이상 나오는 횟수를 출력하자.(정규분포)

dice = [1, 2, 3, 4, 5, 6]

cnt = 0

for i in range(1, 289):
    res = random.choice(dice)
    if res >= 5:
        cnt += 1
print(cnt)
print('확률로 한다면 : ', cnt/1000)
```

```python
#10. 주사위 한개를 288회 던질 때 5 이상의 눈이 나올 횟수를 출력하는데 이 작업을 100번 해서 휫수 100개 출력되게 하시오.

dice = [1, 2, 3, 4, 5, 6]

for j in range(1, 101):
    cnt = 0
    for i in range(1, 289):
        res = random.choice(dice)
        if res >= 5:
            cnt += 1
    print(cnt)
```

```python
#11. 동전 한개와 주사위 한개를 동시에 100번 던져서 / 동적이 앞면이 나오고 주사위의 눈이 5가 나오는 횟수를 
#    출력하는 / 그 작업을 50번 해서 횟수가 50개가 출력되게 하시오. [조건1]
# [조건2] 50번의 횟수를 list객체인 변수 a에 저장한다. 길이도 구하시오.
# [조건3] 조건 2의 값의 평균을 구해라.
# [조건4] 위의 list객체의 값을 하나씩 리턴받아보자. 20 ~ 50숫자만 출력해보자.

coin = ['앞면', '뒷면']
dice = [1,2,3,4,5,6]

a = [] #조건2
for j in range(1, 51):
    cnt = 0
    for i in range(1, 101):
        res = random.choice(dice)
        res2 = random.choice(coin)
        if res == 5 and res2 == '앞면':
            cnt += 1
    a.append(cnt)
print(a, len(a))
print(np.mean(a), np.var(a), np.std(a)) # 조건3

for k in a:
    if k >=20 and k <=50:
        print(k, end = '\t')
```

```python
#15. 서울시 초등학생 백만명의 키를 모집단을 구성하려고 한다.
# [조건 1] 평균 키가 148.5이고 표준편차가 30인 모집단을 만드시오.
# [조건 2] 모집단에서 100명 표본으로 추출하고 100명의 평균키를 리스트객체 a에 입력하는 작업을 10000번 수행한다.
# [조건 3] a객체에 10000개의 표본의 평균키가 대입되도록 하자.

avg = 148.5
std = 30
N = 100000

mogip = np.random.randn(N) * std + avg
mogip


a = []
for i in range(1, 10001):
    sample = np.random.choice(mogip, 100)
    a.append(np.mean(sample))
print(a)
```

```python
# 16. 확률 밀도를 구해보자.
from scipy.stats import norm

avg = 148.5
std = 30
N = 100000

mogip = np.random.randn(N) * std + avg
mogip


a = []
for i in range(1, 10001):
    sample = np.random.choice(mogip, 100)
    a.append(np.mean(sample))
    
x = np.arange(140, 160, 0.001) # 140 ~ 160가지 0.001 간격으로 숫자를 생성
y = norm.pdf(x, np.mean(a), np.std(a))

# 초등학생 키의 표본평균값들에 대한 확률 밀도함수 값이 출력
print(y)

plt.plot(x, y, color = 'blue')
plt.fill_between(x, y, interpolate = True, color = 'skyblue', alpha = 0.7)
plt.show()
```
